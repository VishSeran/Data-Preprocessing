```python
import pandas as pd
import numpy as np
import requests
```

```python
def download(url, fileName):
    response = requests.get(url)
    if response.status_code == 200:
        with open(fileName, "wb") as f:
            f.write( response.content)
```

```python
file_path='https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv'
```

```python
download (file_path, "auto.csv")
file_name = "auto.csv"
```

```python
df = pd.read_csv(file_name)
df.head(20)
```

```python
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
df.columns = headers
df.head(20)
```

```python
### Data Preprocessing
```

```python
### drop the rows if price is NaN or null
```

```python
df.dropna(subset  = ["price"], axis  = 0, inplace = True)
```

```python
df
```

```python
### replace the missing values in normalized-losses column with the column mean

df["normalized-losses"] = pd.to_numeric(df["normalized-losses"], errors = "coerce")
```

```python
print(df["normalized-losses"].dtype)
```

```python
##get missing values
print(df["normalized-losses"].isnull().sum())
df.head(20)
```

```python
#mean value of the column
mean = df["normalized-losses"].mean()
```

```python
##Sign mean value to  missing values
df["normalized-losses"].fillna(mean, inplace = True)
df.head(20)
```

```python
### covert mpg to L/100km (liters per 100km)

df["city-mpg"] = 235/df["city-mpg"]

#rename the city-mpg tp l/100km
df.rename(columns = {"city-mpg" : "L/100km"}, inplace = True)

###convert highway-mpg to L/100km


df.head(20)

```

```python
print(df.isnull().sum())
```

```python

df["highway-mpg"] = 235/df["highway-mpg"]
df.rename(columns = {"highway-mpg" : "highway-L/100km"}, inplace = True)

df.head(20)
```
